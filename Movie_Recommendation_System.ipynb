# **Movie Recommendation System**

### **Recommendation System:**

A recommendation system is a system that filters the content and gives suggestions to user's of their preferences in respective areas. This system can be used in areas like movies, music, products on e-commerce website, articles, news and many more.
It is done using two techniques:-

#### **1. Collaborative Filtering-**

In this approach content is filtered on the basis of past experience of user i.e. what the user has been using or searching and thus content is suggested of of same category.

#### **2. Content-Based Filtering-**

In this approach content is filtered on the basis of definite characteristics and thus products are suggested of matching or similar properties.

**This project will suggest movies to user on the basis of the genre, rating, casts and other factors.**

### **Features Selection**

```python
X=df[["Movie_Genre","Movie_Keywords","Movie_Tagline","Movie_Cast","Movie_Director"]]
```

```python
X.shape
```

```python
X
```

```python
X=X["Movie_Genre"]+" "+X["Movie_Keywords"]+" "+X["Movie_Tagline"]+" "+X["Movie_Cast"]+" "+X["Movie_Director"]
```

```python
X.shape
```

```python
X
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
extractor=TfidfVectorizer()
```

```python
X=extractor.fit_transform(X.values.astype("U"))
```

```python
X.shape
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Match_score=cosine_similarity(X)
```

```python
Match_score
```

```python
Match_score.shape
```

### **Movie name input and suggestion-**

```python
movie_name=input("Enter Movie name: ")
```

```python
movie_list=list(df["Movie_Title"])
```

```python
import difflib
```

```python
Recommendations=difflib.get_close_matches(movie_name,movie_list)
```

```python
print("Closest Movie Recommendation:",Recommendations[0])
index_of_match=df[df["Movie_Title"]==Recommendations[0]]["Movie_ID"].values[0]
print("Movie Id:",index_of_match)
```

### **Enter movie with definite features of your likes-**

```python
movie=input("Enter movie name of definite features(like-genre, cast) you like: ")
```

```python
Recommendations=difflib.get_close_matches(movie,movie_list)
index_of_match=df[df["Movie_Title"]==Recommendations[0]]["Movie_ID"].values[0]
```

```python

recommendation_score=list(enumerate(Match_score[index_of_match]))
```

#### **Top-10 movie recommendation:**

```python
similar_movie_list=sorted(recommendation_score,key=lambda x:x[1],reverse=True)
```

```python
print("Top 10 Recommendations: ")
i=0
while(i<10):
  ind=similar_movie_list[i][0]
  title=df[df.index==ind]["Movie_Title"].values[0]
  print(i+1,"\b.",title)
  i+=1
```
